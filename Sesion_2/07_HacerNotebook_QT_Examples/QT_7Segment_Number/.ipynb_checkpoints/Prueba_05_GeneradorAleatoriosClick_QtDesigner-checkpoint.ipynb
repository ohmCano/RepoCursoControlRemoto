Abrimos el QtDesigner:




Elegimos un **MainWindow** porque es la primera ventana del programa y así debe ser la primera, las siguientes serán del tipo *Dialog*:



Guardamos el archivo recién creado:



Creamos una nueva carpeta para meter el GUI (Graphical User Interface)



Ponemos de la barra lateral izquierda que contiene los Widget, un **Vertical Layout** , que arrastramos desde la izquierda a la ventana recién creada.



Ponemos de la barra lateral izquierda que contiene los Widget, un **LCD Number**, que arrastramos desde la izquierda a la ventana recién creada.



Ponemos de la barra lateral izquierda que contiene los Widget, un **Push Button**, que arrastramos desde la izquierda a la ventana recién creada.



Cambiamos el nombre del **Push Button** por uno que ponga "***Random Number Generator***". Para ello pinchamos 2 veces sombre el nombre actual y lo podremos cambiar.



Vamos a cambiar el fondo de los Números en azul, para ello pinchamos en el rectángulo.



Y ponemos este valor para que el fondo cambie.



Ahora usando la misma ventana de propiedades, vamos a cambiar el fondo del pushbutton a amarillo. Podemos usar el editor de colores:



Podemos seleccionar el objeto del PushButton y cambiarle las propiedades o seleccionar la LAYER y cambiar ambos a la vez.



Ahora vamos a cambiar el tipo de letra del texto del PUSHBUTTON. Para ello, pinchamos sobre el objeto botón y  en la ventana de la derecha miramos en el **OBJECT INSPECTOR** y miramos el apartado font.



Vamos a poner tamaño del Tipo de Letra 14 en un TTF que se llama MS Sheell Diag 2 , 14 puntos.



Activamos las negritas para que se vea bien el text dentro del ***PushButoom***.

Pinchando sobre el fondo de la aplicación pero fuera de objetos, le damos al RMB y al final de la ventana emergente  nos vamos a *Lay Out Horizantally*.



Y se quedan mas centradito y bonito



Ahora podemos cambiar el tamaño de la ventana que todo queda bien ajustado (RESPONSIVE en sentido amplio).

Vamos a poner un título a la ventana de la aplicación, para ello buscamos el parámetro **Title**

 

 y guardamos el proyecto de GUI dándole a FILE.> Save

 

 Para poner un icono al programa:  Podemos buscar muchas imágenes en este portal:  https://iconos8.es/

 

```python
#Ejemplo muy similar al sacado de: https://nitratine.net/blog/post/how-to-import-a-pyqt5-ui-file-in-a-python-gui/


from PyQt5 import QtWidgets, uic
import sys

#importamos la función generadora del randint: https://www.w3schools.com/python/ref_random_randint.asp
from random import randint

class MainWindow(QtWidgets.QMainWindow):

    #Mainwindow class constructor
    def __init__(self):
        #Call MainWindow super class from PyQt and load the design created in Qt-Designer.
        super(MainWindow, self).__init__()
        uic.loadUi(".\GUI\Gui_Qt_MainWindow.ui",self)
        #Para poner un icono en QtDesigner usamos: https://iconos8.es/

        #Vamos a anclar Evento en el Button con método
        self.button_generator = self.findChild(QtWidgets.QPushButton, 'pushButton') # Find the button
        self.button_generator.clicked.connect(self.rand_generator_pressed) # Remember to pass the definition/method, not the return value!

    def rand_generator_pressed(self):
        random_num=randint(1,300)
        self.lcdNumber.display(random_num)

if __name__ == "__main__":
    
    App= QtWidgets.QApplication(sys.argv)
    window=MainWindow()
    window.show()
    sys.exit(App.exec())
```
