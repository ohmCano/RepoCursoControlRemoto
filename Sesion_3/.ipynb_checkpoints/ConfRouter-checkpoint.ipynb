## Paso 2: Conexión a la LAN del router con cable Ethernet

NOTA: En PC nuevos puede no haber interfaz ETH. Lo que selen tener es un interfaz USB C, por lo que es necesario un adaptador ESB C <--> ETH

En el administrador de dispositivos nos aparece el adaptador:



En el centro de redes y recursos compartidos nos aparece una nueva interfaz de red cableada ETH. Entramos en esta interfaz (doble clic) --> propiedades --> doble click en TCP/IPV4. Esto abré la ventana que se muestra a continuación. Aquí nos aseguramos de que esta interfaz está a la espera de recibir una IP automáticamente.

En el router reseteado, el servicio DHCP (Dynamic Host Control Protocol) habilitado en la LAN se encargará de asignar una IP a nuestro PC. Otra opción sería condigurar una IP fija en el equipo dentro del rango de dirección IP que reparte el servicio DHCP del router. 



Para saber la IP que nos ha asignado el router y la puerta de enlace predeterminada abrimos una terminal y escribimos el comando ipconfig



A continuación, escribimos la dirección de la puerta de enlace predeterminada en el navegador y el servidor web habilitado en la LAN del router servirá una página web en la que podremos configurar el router
